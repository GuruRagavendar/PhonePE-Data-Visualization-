```python
import pandas as pd
import json
import os
from pprint import pprint 
from sqlalchemy import create_engine
```

Aggregated Transaction

```python
#This is to direct the path to get the data as states

path_agg_trans= r"C:/Users/mathe/Documents/guvi projects/PhonePe/pulse-master/pulse-master/data/aggregated/transaction/country/india/state/"
Agg_state_list=os.listdir(path_agg_trans)
Agg_state_list

#Agg_state_list--> to get the list of states in India

#<------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------>#

#This is to extract the data's to create a dataframe

clm={'State':[], 'Year':[],'Quater':[],'Transacion_type':[], 'Transacion_count':[], 'Transacion_amount':[]}

for i in Agg_state_list:
    p_i=path_agg_trans+i+"/"
    Agg_yr=os.listdir(p_i)
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['transactionData']:
              Name=z['name']
              count=z['paymentInstruments'][0]['count']
              amount=z['paymentInstruments'][0]['amount']
              clm['Transacion_type'].append(Name)
              clm['Transacion_count'].append(count)
              clm['Transacion_amount'].append(amount)
              clm['State'].append(i.title().replace('-',' '))
              clm['Year'].append(int(j))
              clm['Quater'].append(int(k.strip('.json')))

# Succesfully created a dataframe
Agg_Trans=pd.DataFrame(clm)
# Agg_Trans.isnull().sum()
# # Agg_Trans.info()
Agg_Trans
```

Aggregated User

```python
#This is to direct the path to get the data as states

path= r"C:/Users/mathe/Documents/guvi projects/PhonePe/pulse-master/pulse-master/data/aggregated/user/country/india/state/"
Agg_state_list=os.listdir(path)
#Agg_state_list--> to get the list of states in India

#<------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------>#

#This is to extract the data's to create a dataframe

clm={'State':[], 'Year':[],'Quater':[],'Brand':[], 'User_count':[], 'User_percentage':[]}

for i in Agg_state_list:
    p_i=path+i+"/"
    Agg_yr=os.listdir(p_i)
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            try:
              for z in D['data']['usersByDevice']:
                B_Name= z['brand']
                count= z['count']
                percentage= z['percentage']
                clm['Brand'].append(B_Name)
                clm['User_count'].append(count)
                clm['User_percentage'].append(percentage)
                clm['State'].append(i.title().replace('-',' '))
                clm['Year'].append(int(j))
                clm['Quater'].append(int(k.strip('.json')))
            except : pass

# Succesfully created a dataframe
Agg_Users=pd.DataFrame(clm)
Agg_Users.isnull().sum()
Agg_Users.info()
Agg_Users
```

```python
#This is to direct the path to get the data as states

path= r"C:/Users/mathe/Documents/guvi projects/PhonePe/pulse-master/pulse-master/data/map/transaction/hover/country/india/state/"
Map_state_list=os.listdir(path)
#Map_state_list--> to get the list of states in India

#<------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------>#

#This is to extract the data's to create a dataframe

clm={'State':[], 'Year':[],'Quater':[],'Amount':[], 'Transaction_count':[], 'District':[]}

for i in Map_state_list:
    p_i=path+i+"/"
    Map_yr=os.listdir(p_i)
    for j in Map_yr:
        p_j=p_i+j+"/"
        Map_yr_list=os.listdir(p_j)
        for k in Map_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            try:
              for z in D['data']['hoverDataList']:
                amount= z['metric'][0]['amount']
                count= z['metric'][0]['count']
                district= z['name']
                clm['Amount'].append(amount)
                clm['Transaction_count'].append(count)
                clm['District'].append(district.title().replace(' District',''))
                clm['State'].append(i.title().replace('-',' '))
                clm['Year'].append(int(j))
                clm['Quater'].append(int(k.strip('.json')))
            except : pass

 # Succesfully created a dataframe
Map_Trans=pd.DataFrame(clm)
Map_Trans.isnull().sum()
Map_Trans.info()
Map_Trans
```
